## **DROPOUT REGULARIZATION EXAMPLE**

IMPORTING DEPENDENCIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

LOADING AND PREPROCESSING DATA

```python
df = pd.read_csv("/content/sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.shape
```

```python
# check for nan values
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts() # label is not skewed
#111 MINES
#97 ROCKS
```

```python
#X: 60th columns is the label column so drop it
#y: 60th column i.e. label column
X = df.drop(60, axis='columns')
y = df[60]
y.head()
```

```python
X
```

one hot encoding

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0
```

```python
y.value_counts()
```

```python
X.head()
```

TRAIN-TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
X_train.head()
```

MODEL WITHOUT DROPOUT

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),  #input nodes: 60, hidden layer 1: connected to input layer
    keras.layers.Dense(30, activation='relu'),                #hidden layer2: 30 nodes
    keras.layers.Dense(15, activation='relu'),                #hidden layer3: 15 nodes
    keras.layers.Dense(1, activation='sigmoid')               #output layer: 1 node
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

WITH DROPOUT LAYER

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
